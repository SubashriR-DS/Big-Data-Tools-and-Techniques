```
# create an RDD -2023
rddFile =sc.textFile("FileStore/tables/clinicaltrial_2023.csv")
```

```
# delimiters
rdd1 = rddFile.map(lambda x: x.strip(',').strip('\"')).map(lambda x: x.split('\t'))
rdd1.take(5)
```

```
pharma = sc.textFile("FileStore/tables/pharma.csv")
```

```
pharma_rdd= pharma.map(lambda x: x.strip(',').strip('\"').rstrip('\"').split(','))
pharma_rdd.collect()
```

```
pharma_rdd1 = pharma_rdd.map(lambda x: x[1].replace('"', ''))
pharma_rdd1.collect()
```

```
headerrdd1 =rdd1.first()
rdd2 =rdd1.filter(lambda x: x != headerrdd1)  # remove the header 
rdd2.take(10)
```

```
# Q1.
rdd3= rdd2.map(lambda x: x[0]).distinct()
count_studies=rdd3.collect()
len(count_studies)
```

```
# Q2.
type_column = rdd1.first().index('Type')
rdd1.filter(lambda x: len(x) > type_column + 1).map(lambda x : (x[type_column], 1)).filter(lambda row : row[0] != 'Type').reduceByKey(lambda a,b :a + b).filter (lambda x:x[0] != '').sortBy(lambda x :x[1],ascending =False).collect()
```

```
# Q3.
condition_column = rdd1.first().index('Conditions')

rdd1.flatMap(lambda x: x[condition_column].split('|')).filter(lambda row: row != 'Conditions').filter(lambda row: row != '').map(lambda x: (x, 1)).reduceByKey(lambda a,b: a + b).sortBy(lambda x: x[1], ascending=False).take(5)
```

```
# Q4.
sponsor_column =rdd1.first().index('Sponsor')

rdd1.map(lambda x: x[sponsor_column]).filter(lambda row: row != 'Sponsor').subtract(pharma_rdd.map(lambda x: x[1].replace('"', ''))).map(lambda x: (x, 1)).reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], ascending=False).take(12)
```

```
# Q5.

rdd1 = rdd1.map(lambda x: [element.replace(',', '').replace('"', '') for element in x])

month_selector = {"01": "Jan", "02":"Feb", "03":"Mar", "04":"Apr", "05":"May", "06":"Jun", "07":"Jul", "08":"Aug", "09":"Sept", "10":"Oct", "11":"Nov", "12":"Dec"}

completion_column = rdd1.first().index('Completion')
status_column = rdd1.first().index('Status')

rdd1.filter(lambda x: len(x) > type_column + 1).map(lambda x: (x[completion_column], x[status_column])).filter(lambda x: x[0] != 'Completion').filter(lambda x: x[1] == 'COMPLETED' or x[1] == 'Completed').map(lambda x: (x[0][5:7], x[0][0:4])).filter(lambda x: x[1] == "2023").map(lambda x: (x[0], 1)).reduceByKey(lambda a,b: a + b).map(lambda x: (month_selector[x[0]], x[1])).collect()
```

```
import matplotlib.pyplot as plt

# Data
data = [('Febuary', 1272), ('October', 1058), ('June', 1619), ('January', 1494), ('April', 1324), ('August', 1230), ('May', 1415), ('July', 1360), ('November', 909), ('September', 1152), ('December', 1082), ('March', 1552)]

# Extracting months and counts
months = [item[0] for item in data]
counts = [item[1] for item in data]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(months, counts, color='purple')
plt.title('Clinical Trial Completion Count by Month in 2023')
plt.xlabel('Months')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
